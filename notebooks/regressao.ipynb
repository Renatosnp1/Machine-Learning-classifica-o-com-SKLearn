```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# 1. Gerar um conjunto de dados linear simples
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # Geração da relação linear com ruído

# 2. Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Treinar um modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# 4. Prever valores para os dados de teste
y_pred = model.predict(X_test)

# 5. Plotar os dados de treino e a linha de regressão
plt.scatter(X_train, y_train, color='blue', label='Dados de Treino')
plt.scatter(X_test, y_test, color='green', label='Dados de Teste')

# 6. Traçar a linha de regressão
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_range_pred = model.predict(X_range)
plt.plot(X_range, y_range_pred, color='red', linewidth=2, label='Linha de Regressão')

# 7. Personalizar o gráfico
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.title('Modelo de Regressão Linear')
plt.legend()
plt.show()
```

```python
plt.scatter(X_test, y_test - y_pred, color='purple', label='Resíduos')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Resíduos da Regressão Linear')
plt.xlabel('Feature 1')
plt.ylabel('Resíduo')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# 1. Gerar um conjunto de dados linearmente separável
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                            n_clusters_per_class=1, n_samples=100, random_state=42)

# 2. Treinar o modelo de regressão logística
model = LogisticRegression()
model.fit(X, y)

# 3. Criar uma grade de pontos para visualizar os limites de decisão
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                    np.arange(y_min, y_max, 0.02))

# 4. Prever as classes para cada ponto da grade
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 5. Plotar os limites de decisão
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)

# 6. Plotar os pontos de dados
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=100, cmap=plt.cm.coolwarm)

# 7. Personalizar o gráfico
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Limites de Decisão usando Regressão Logística')
plt.show()
```
